### Anthropic SDK ã¨ Messages API
ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«å…¨ä½“ã‚’é€šã—ã¦ã€[Anthropic python SDK](https://docs.anthropic.com/claude/reference/client-sdks)ã¨[Messages API](https://docs.anthropic.com/claude/reference/messages_post)ã‚’ä½¿ç”¨ã—ã¾ã™ã€‚

ä»¥ä¸‹ã¯ã€ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«ã§ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’å®Ÿè¡Œã™ã‚‹éš›ã®ä¾‹ã§ã™ã€‚ã¾ãšã€ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’Claudeã«é€ä¿¡ã—ã€ClaudeãŒç”Ÿæˆã—ãŸå›ç­”ã‚’è¿”ã™ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã§ã‚ã‚‹`get_completion`ã‚’ä½œæˆã—ã¾ã™ã€‚ä»Šã™ããã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚

ã¾ãšã€ãƒ¢ãƒ‡ãƒ«åã¨ãƒªãƒ¼ã‚¸ãƒ§ãƒ³ã‚’è¨­å®šã—ã¦ä¿å­˜ã—ã¾ã™ã€‚

```python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
```

æ¬¡ã«ã€ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’Claudeã«é€ä¿¡ã—ã€ClaudeãŒç”Ÿæˆã—ãŸå›ç­”ã‚’è¿”ã™ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã§ã‚ã‚‹`get_completion`ã‚’ä½œæˆã—ã¾ã™ã€‚ä»Šã™ããã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚

```python
import boto3
import json

bedrock = boto3.client('bedrock-runtime',region_name=AWS_REGION)

def get_completion(prompt):
    body = json.dumps(
        {
            "anthropic_version": '',
            "max_tokens": 2000,
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.0,
            "top_p": 1,
            "system": ''
        }
    )
    response = bedrock.invoke_model(body=body, modelId=MODEL_NAME)
    response_body = json.loads(response.get('body').read())

    return response_body.get('content')[0].get('text')
```

ãã‚Œã§ã¯ã€Claudeã«å‘ã‘ã¦ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’æ›¸ãã€`get_completion`ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã‚’å®Ÿè¡Œã—ã¦Claudeã®å‡ºåŠ›ã‚’è¡¨ç¤ºã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚ä»¥ä¸‹ã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã™ã‚‹ã¨ã€ãã®ä¸‹ã«Claudeã‹ã‚‰ã®å›ç­”ãŒè¡¨ç¤ºã•ã‚Œã¾ã™ã€‚

Claudeã‹ã‚‰ç•°ãªã‚‹å›ç­”ã‚’å¼•ãå‡ºã™ãŸã‚ã«ã€ãŠæ°—è»½ã«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆæ–‡å­—åˆ—ã‚’å¤‰æ›´ã—ã¦è©¦ã—ã¦ã¿ã¦ãã ã•ã„ã€‚

```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```
